# BlackRock Data Analyzing 20170708

Library

```python
import pandas as pd
import numpy as np
import scipy as sp
import math
from datetime import date
from datetime import datetime

import matplotlib as mpl
import matplotlib.pyplot as plt
```

Reading Data

```python
df=pd.read_csv("BlackRock20170708.csv")
df.head()
```

```python
date=df["Date"]
df["DDate"]=pd.to_datetime(date)
df['Year'] = df['DDate'].dt.year
df.head()
```

```python
x=df["PX_MID"]
y=df["FNCL 5.5"]
labels = df['Year']
df = pd.DataFrame(dict(x=x, y=y, label=labels))

groups = df.groupby('label')

# Plot
plt.rcParams.update(pd.tools.plotting.mpl_stylesheet)
colors = pd.tools.plotting._get_standard_colors(len(groups), color_type='random')

fig, ax = plt.subplots(figsize=(16, 12))
ax.set_color_cycle(colors)
ax.margins(0.05)
for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=3, label=name)
ax.legend(numpoints=1)

plt.show()
```

```python
yeargp=df.groupby(df["Year"])[["FNCL 5.5","PX_MID"]].std()
yeargp["relatvestd"]=yeargp["FNCL 5.5"]/yeargp["PX_MID"]
print(yeargp)
```
